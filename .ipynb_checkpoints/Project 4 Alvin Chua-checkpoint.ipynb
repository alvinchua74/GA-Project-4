# Executive Summary

In this project, I wanted to find out how job postings and the relevant details affect salary, and also whether components of a job posting distinguish jobs from each other.

In order to scope the project to a feasible level, I only search for job postings within Singapore, and data-related jobs (for e.g. data scientist, data analyst, data engineer etc). The relevant job postings are then done via web scrapping from the Singapore job portal [MyCareersFuture](https://www.mycareersfuture.sg/).

## Prior Steps

The first step of the project is to clean the data and look for any interesting features to engineer/choose for the modelling part later: 
1. I wanted to find out whether the locations of the offices will impact salary. I first used the postal codes of the companies and grouped them into district zones, and regions (north, south, east, west etc) thereafter.  
2. As the job titles of posting vary by a large extent, I had to group the titles into job categories for easier referencing too. The job categories may indicate a certain degree of salary prediction.
3. Some job postings have numerous selections of seniority levels for one job position, so some engineering is done to distinguish whether a certain seniority level is indicated in the posting. This step is important, as the seniority level usually impacts the salary scale by a huge extent.
4. Employment types are important too, as permanent staff may be paid higher than contract staff, for example. Hence I also engineered the feature to check whether a certain employment type is indicated in the job posting.
5. Some industries are known to pay more than others, hence this feature was also engineered to check whether a certain industry is indicated in the job posting. There will also be instances of multiple industries in each job posting.
6. As there is a large number of text from the description and requirements of each job, I applied Natural Language Processing tools to change them into useful features.
7. Finally, as most of the salaries quoted have a lower and upper range, I used the average of each job posting as a combined feature instead.



## Part 1

The second step is to look for strong indicators that impact salary. I framed this problem into a classification problem, where a high-paying job is defined with a salary of \\$6,500 or more, while a low-paying job is defined with a salary of less than \\$6,500. The cut-off value of \\$6,500 is chosen as the median salary in the salary is as such. This will in turn become my target variable for this part of the problem.

Three different classification models, from simple models (for easier interpretability) to complex models (for optimising performance) were used and compared for indicators of strong features. Each model had slightly different results, but overall features like seniority levels highly impact salary. Some words from the description and requirements section (for e.g. educational requirements, skils, industries)highly impact salary too.


## Part 2

In the second part, I will like to find out whether certain features distinguish job categories. Hence, the job categories that I have engineered will be my target variable for this part of the problem. Some minor tweaking is done, where I will only want to distinguish whether a job posting is for a  data scientist role or not.

I only used the tokenised words from the job description and requirements sections as feature for this classification problem. As per Part 1, I also used the same three models to compare the results. Although their results differ slightly, some strong indicators appear in at least two of the models' results, which highly suggest its importance.

Word like "phd", "data science", "deep learning" frequently appear as strong indicators in at least two of the three models. Interesting words like "mathematics" also appear as important features. Those words indicate that companies are more likely to hire data scientists who hold a PhD and has experience in deep learning. Maybe someone who majored in Mathematics?

## Part 3

In the third and last part, I will like to solve the issue of telling a prospective client incorrectly that they would get a high salary job. It makes more sense to tell clients incorrectly that they would get a low salary job than a high salary job. 

I chose one of the tuned models (logistice regression) from Part 2 for this purpose. As the standard threshold for binary classification problems is a probability of 0.5 to separate different classes, I will have to increase this threshold to a probability of 0.8 for this purpose. In that case, I ended up with 0 wrongly classified high-paying jobs. 

However, the move decreased the number of rightly classified high-paying jobs, and increased the number of wrongly classified low-paying jobs. Nevertheless, the move is still sound, as we rather tell clients incorrectly that the pay is low rather than high.

## Limitations


One limitation I faced was that there were too many words that could be treated as stop words, but it was a tedious process to expand the stop words library. 

A possible risk I faced was to treat the words that appear as binary counts instead of actual counts. In my opinion, so long as a keyword appear in a document for this context, it may be an indicator of salary or job category. Using the actual number of words that appear in one document may not make much sense. Hence I proceeded with the binary method.

Another risk was that I combined the upper and lower salary quotes as a single average salary amount per job posting. In reality, it may not be that straightforward. Junior positions tend to have a narrower salary range, and senior positions tend to have a wider salary range. Companies could also have different styles of budgeting/strategic planning for the employees' salaries, so some companies may tend to pay the final salary amount nearer to either the lower or the upper salary bound.

```python
# for each job title in the list of titles
for title in titles:
    
    #open the chrome browser and go to base url (MyCareersFuture)
    #and let it sleep for a while to before scrapping.
    #I used selenium
    driver1 = webdriver.Chrome(executable_path="C:\chromedriver.exe")
    driver1.get(base_url)
    sleep(random.randint(4,6))
    
    #search for search bar
    elem = driver1.find_element_by_name("search-text")
    #clear search bar
    elem.clear()
    #type job title using selenium
    elem.send_keys(title)
    #press enter in search bar
    elem.send_keys(Keys.RETURN)
    sleep(random.randint(4,6))
    
    #get current page url so that I can edit the search results page
    #number directly on the url
    current_page_url = driver1.current_url

    html_search = driver1.page_source
    #find total number of jobs per search
    numjobs = int(Selector(text=html_search).xpath('//span[@id="search-results-message"]/div/text()').extract()[0].split(' ')[0])
    
    driver2 = webdriver.Chrome(executable_path="C:\chromedriver.exe")
    driver3 = webdriver.Chrome(executable_path="C:\chromedriver.exe")
    driver4 = webdriver.Chrome(executable_path="C:\chromedriver.exe")
    driver5 = webdriver.Chrome(executable_path="C:\chromedriver.exe")
    
    drivers = [driver1,driver2,driver3,driver4,driver5]
    
    # each search page gives 20 results, so I find the quotient of the 
    # total number of jobs searched divided by 20.
    # in the url, the first search results page has a 'page0',
    # so I used the range function to start from 0 until the last page
    # and edit the url directly
    
    
    
    print('Scraping',title,'jobs:')
    for i in range(0,int(divmod(numjobs,20)[0])+1):
        
        #new url link for search page
        new_url = current_page_url[:-1]+str(i)
        driver1.get(new_url)
        sleep(random.randint(4,6))
        print('Scraping page',i,'......')
        html_search = driver1.page_source
        #capture job posting links' URL for each page
        posting_links = Selector(text=html_search).xpath('//a[contains(@href,"/job/")]/@href').extract()

        #
        for index in range(0,len(posting_links),5):
            
            try:
                driver1.get(base_url+posting_links[index])
            except:
                driver1.close()
                
            try:    
                driver2.get(base_url+posting_links[index+1])
            except:
                driver2.close()
            
            try:
                driver3.get(base_url+posting_links[index+2])
            except:
                driver3.close()
                
            try:    
                driver4.get(base_url+posting_links[index+3])
            except:
                driver4.close()
            
            try:
                driver5.get(base_url+posting_links[index+4])
            except:
                driver5.close()
                
            sleep(random.randint(4,7))
            
            # this is the link for each individual job posting page
            

            # go to page
            ##driver.get(job_listing_url)
            
            for driver in drivers:
            
                try:
                    html_job = driver.page_source
                except:
                    continue
                # append all relevant info, only when there is an error then we
                # append 'NA'
                try:
                    job_coyname = Selector(text=html_job).xpath('//p[@name="company"]/text()').extract()[0]
                    coyname_list.append(job_coyname)
                except:
                    coyname_list.append('NA')

                try:
                    job_title = Selector(text=html_job).xpath('//h1[@id="job_title"]/text()').extract()[0]
                    title_list.append(job_title)
                except:
                    title_list.append('NA')

                try:
                    job_id = Selector(text=html_job).xpath('//span[@data-cy="jobinfo__jobpostid--span"]/text()').extract()[0]
                    id_list.append(job_id)
                except:
                    id_list.append('NA')

                try:
                    job_add = Selector(text=html_job).xpath('//p[@id="address"]/a/text()').extract()[0]
                    add_list.append(job_add)
                except:
                    add_list.append('NA')

                try:
                    job_emptype =  Selector(text=html_job).xpath('//p[@id="employment_type"]/text()').extract()[0]
                    emptype_list.append(job_emptype)
                except:
                    emptype_list.append('NA')

                try:
                    job_seniority = Selector(text=html_job).xpath('//p[@id="seniority"]/text()').extract()[0]
                    seniority_list.append(job_seniority)
                except:
                    seniority_list.append('NA')

                try:
                    job_minexp = Selector(text=html_job).xpath('//p[@id="min_experience"]/text()').extract()[0]
                    minexp_list.append(job_minexp)
                except:
                    minexp_list.append('NA')

                try:
                    job_cat = Selector(text=html_job).xpath('//p[@id="job-categories"]/text()').extract()[0]
                    cat_list.append(job_cat)
                except:
                    cat_list.append('NA')

                try:
                    job_salarylower = Selector(text=html_job).xpath('//span[@class="dib"]/text()').extract()[0]
                    salarylower_list.append(job_salarylower)
                except:
                    salarylower_list.append('NA')

                try:
                    job_salaryupper = Selector(text=html_job).xpath('//span[@class="dib"]/text()').extract()[1]
                    salaryupper_list.append(job_salaryupper)
                except:
                    salaryupper_list.append('NA')

                try:
                    job_salarytype = Selector(text=html_job).xpath('//span[@class="salary_type dib f5 fw4 black-60 pr1 i pb"]/text()').extract()[0]
                    salarytype_list.append(job_salarytype)
                except:
                    salarytype_list.append('NA')

                try:
                    job_description = Selector(text=html_job).xpath('//div[@id="description-content"]//text()').extract()
                    description_list.append(job_description)
                except:
                    description_list.append('NA')

                try:
                    job_requirements = Selector(text=html_job).xpath('//div[@id="requirements-content"]//text()').extract()
                    requirements_list.append(job_requirements)
                except:
                    requirements_list.append('NA')

                try:
                    job_skills = Selector(text=html_job).xpath('//label[@class="br-pill ba pv1 pl2 pr3 pointer PillButton__pillbutton___1JvIO"]/text()').extract()
                    skills_list.append(job_skills)
                except:
                    skills_list.append('NA')   

        #to check that each search results page is scraped
        print('Scraping page',i,'done!')
    #to check that scraping is done for each job title     
    print('Scraping for',title,'jobs done!')
    print('-'*30)
    for driver in drivers:
        driver.quit()
```

```python
def replace_chars(string):
    """
    This function takes in a string as an argument and
    removes the characters
    """
    clean_string = string.replace('[','')\
                        .replace(']','')\
                        .replace("'","")\
                        .replace('"','')\
                        .replace('&','and')\
                        .replace('•','')\
                        .replace('\\xa0',' ')\
                        .replace('*','')\
                        .replace('~','')\
                        .replace(' \\t','')\
                        .replace("\\",'')\
                        .replace(',  t,','')\
                        .replace(',   ','')\
                        .replace('  ,  , ','')
    clean_string = clean_string.strip()
                        
    return clean_string
```

```python
# create dictionaries of zones and district mapping 
# as per picture and link above

zones = {
    'central' : [1,2,3,4,6,7,8,9,10,11,12,13,14,21,26],
    'east' : [15,16,17,18],
    'west' : [5,22,23],
    'northeast' : [19,20,28],
    'north' : [24,25,27]
}

postalcode_to_district = {
    1:['01','02','03','04','05','06'],
    2:['07','08'],
    3:['14','15','16'],
    4:['09','10'],
    5:['11','12','13'],
    6:['17'],
    7:['18','19'],
    8:['20','21'],
    9:['22','23'],
    10:['24','25','26','27'],
    11:['28','29','30'],
    12:['31','32','33'],
    13:['34','35','36','37'],
    14:['38','39','40','41'],
    15:['42','43','44','45'],
    16:['46','47','48'],
    17:['49','50','81'],
    18:['51','52'],
    19:['53','54','55','82'],
    20:['56','57'],
    21:['58','59'],
    22:['60','61','62','63','64'],
    23:['65','66','67','68'],
    24:['69','70','71'],
    25:['72','73'],
    26:['77','78'],
    27:['75','76'],
    28:['79','80']
}
```

```python
districtcodes = []

#change 1st two numbers of postal code to district code
for postcode in jobs.postal_code:
    for district,postalcode in postalcode_to_district.items():
        if postcode in postalcode:
            districtcodes.append(district)
    if postcode=='NA':
        districtcodes.append('NA')

regions = []

#change district code to region
for code in districtcodes:
    for region,districtcode in zones.items():
        if code in districtcode:
            regions.append(region)
    if code=='NA':
        regions.append('NA')

#create new column in dataframe
jobs['region'] = regions

#drop postal code column
jobs.drop('postal_code',axis=1,inplace=True)
#drop address
jobs.drop('add',axis=1,inplace=True)

```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
from scipy.sparse import csr_matrix, hstack
from sklearn.model_selection import train_test_split,cross_val_score,GridSearchCV, KFold, StratifiedKFold
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, precision_recall_fscore_support
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB, BernoulliNB
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
```

```python
#do train-test split, by stratifying the target variable

y = jobs.is_highsalary
X = jobs.drop('is_highsalary',axis=1)

X_train, X_test, y_train, y_test = train_test_split(
    X,y,test_size=0.2,stratify=y,random_state=42
)
```

```python
#get baseline score first
y_test.mean()
```

```python
#create various models to check scores using above function
logreg = LogisticRegression(random_state=42)
rfc = RandomForestClassifier(random_state=42)
gbc = GradientBoostingClassifier(random_state=42)

models = [logreg,rfc,gbc]
```

```python
#define a function that helps to compare cross-validated scores of models
def models_score(models,X_train,y_train,scoring):
    scores_dict = {}
    
    for model in models:
        scores = cross_val_score(model,
                                 X_train,
                                 y_train,
                                 cv=5,
                                verbose=1,
                                n_jobs=-1,scoring=scoring)
        scores_dict[model] = scores
    
    df = pd.DataFrame(scores_dict)
    df.columns = ['Logistic Regression','Random Forest','Gradient Boosting Classifier']
    
    return df
```

```python
models_scoring = models_score(models,X_train,y_train,'roc_auc')
models_scoring.describe()
```

```python
plt.figure(figsize=(15,10))
models_scoring.boxplot()
plt.show()
```

From the above statistics and boxplots, it seems like the gradient boosting classifier is performing a little better. Nevertheless, all 3 models have roughly the same mean cross-validated score, and small standard deviation of the scores.

```python
#define a function to show various classification metrics
def classification_metrics(models,X_train,X_test,y_train,y_test):
    for model in models:
        model = model.fit(X_train,y_train)
        y_pred = model.predict(X_test)
        y_pred_proba = model.predict_proba(X_test)
        print(str(model).split('(')[0])
        print(confusion_matrix(y_test,y_pred))
        print()
        print(classification_report(y_test,y_pred))
        print()
        print('ROC AUC score:',roc_auc_score(y_test,y_pred_proba[:,1]))
        print('-'*50)
```

```python
classification_metrics(models,X_train,X_test,y_train,y_test)
```

    Finetuning hyperparameters

```python
#logistic regression with grid search
def logreg_with_gs(X,y):
    
    logreg = LogisticRegression(random_state=42)
    
    logreg_params = {
        'penalty': ['l1','l2'],
        'C': np.logspace(-3,3,100),
        'solver':['liblinear']
    }
    
    logreg_grid = GridSearchCV(logreg,
                               param_grid=logreg_params,
                              cv=5,
                              verbose=1,
                              n_jobs=-1)
    logreg_grid.fit(X,y)
    
    print('Best score:',logreg_grid.best_score_)
    print('Best parameters',logreg_grid.best_params_)
    
    return logreg_grid.best_estimator_
```

```python
logreg_with_gs(X_kbest_train,y_train)
```

```python
#random forest with grid search
def rfc_with_gs(X,y):
    
    rfc = RandomForestClassifier(random_state=42)
    
    rfc_params = {
        'n_estimators':[10,100,1000],
        'min_samples_split':range(2,8),
        'max_features': [1, 2, 3]
    }
    
    rfc_grid = GridSearchCV(rfc,
                               param_grid=rfc_params,
                              cv=5,
                              verbose=1,
                              n_jobs=-1)
    rfc_grid.fit(X,y)
    
    print('Best score:',rfc_grid.best_score_)
    print('Best parameters',rfc_grid.best_params_)
    
    return rfc_grid.best_estimator_
```

```python
rfc_with_gs(X_kbest_train,y_train)
```

```python
#gradient boosting with grid search
def gbc_with_gs(X,y):
    
    gbc = GradientBoostingClassifier(random_state=42)
    
    gbc_params = {
        'n_estimators':[10,100,1000],
        'learning_rate':[0.1,0.2,0.5,1],
        'min_samples_split':range(2,8),
        'max_features': [1, 2, 3]
    }
    
    gbc_grid = GridSearchCV(gbc,
                               param_grid=gbc_params,
                              cv=5,
                              verbose=1,
                              n_jobs=-1)
    gbc_grid.fit(X,y)
    
    print('Best score:',gbc_grid.best_score_)
    print('Best parameters',gbc_grid.best_params_)
    
    return gbc_grid.best_estimator_
```

```python
#define a function that helps to compare cross-validated scores of models
#scores are ROC AUC
#for imbalanced class problem
def models_score_imb(models,X_train,y_train):
    scores_dict = {}
    
    for model in models:
        
        skf = StratifiedKFold(n_splits=5,random_state=42)
        
        scores = []

        for train_index, test_index in skf.split(X_train,y_train):
            
            Xfold_train, Xfold_test = X_train.iloc[train_index,:], X_train.iloc[test_index,:]
            yfold_train, yfold_test = y_train.iloc[train_index], y_train.iloc[test_index]

            #SMOTE oversampling tool
            sm = SMOTE(ratio = 'minority', k_neighbors=20, random_state=42)
            X_smote, y_smote = sm.fit_sample(Xfold_train, yfold_train)

            model.fit(X_smote, y_smote)
            yfold_pred = model.predict_proba(Xfold_test)[:,1]
            score = roc_auc_score(yfold_test,yfold_pred)
            scores.append(score)
        
        scores_dict[model] = scores
    
    df = pd.DataFrame(scores_dict)
    df.columns = ['Logistic Regression','Random Forest','Gradient Boosting Classifier']
    
    return df
```

```python
#create various models to check scores using above function
logreg_new = LogisticRegression(random_state=42)
rfc_new = RandomForestClassifier(random_state=42)
gbc_new = GradientBoostingClassifier(random_state=42)

models_new = [logreg_new,rfc_new,gbc_new]
```

# 3. Try to mitigate problem of over-promising client that quoted salary is high

I chose to adjust the logistic regression model that was modelled in Part 2 to tackle this problem.

Since the "boss" would rather tell a client incorectly that they would get a lower salary job than tell a client incorrectly that they would get a high salary job, we need to tweak the original threshold which is a probability of 0.5 to something else.

In this context, we need to decrease the false positives to 0.
I will use the probabilities predicted and increase the threshold (default is probability of 0.5) of a high salary  to 0.8.

```python
Y_pp = pd.DataFrame(logreg_grid_new.predict_proba(Xnew_test),columns=['class_0_pp','class_1_pp'])
#decrease false positives to 0, so
Y_pp['pred_class_thresh80'] = [1 if x >= 0.8 else 0 for x in Y_pp.class_1_pp.values]
Y_pp.head(10)
```

```python
conmat = np.array(confusion_matrix(ynew_test, Y_pp.pred_class_thresh80.values, labels=[0,1]))
conmat
confusion = pd.DataFrame(conmat, index=['is_lowsalary', 'is_highsalary'],
                        columns=['predicted_lowsalary','predicted_highsalary'])
confusion
```

From the confusion matrix, it seems that the adjusted model will not predict any high salary job incorrectly. However, due to this action, my true postives have decreased and my false negatives have increased.

Nevertheless, the results are still okay from the point of view of the recruitment firm. The company would rather have more false negatives that a single count of false positive.

```python
from sklearn.metrics import roc_curve, auc
```

```python
# For class 1, find the area under the curve.
fpr, tpr, _ = roc_curve(ynew_test, Y_pp.pred_class_thresh80)
roc_auc = auc(fpr, tpr)

# Plot of a ROC curve for class 1 (is_highsalary).
plt.figure(figsize=[8,8])
plt.plot(fpr, tpr, label='ROC curve (area = %0.3f)' % roc_auc, linewidth=4)
plt.plot([0, 1], [0, 1], 'k--', linewidth=4)
plt.xlim([-0.05, 1.0])
plt.ylim([-0.05, 1.05])
plt.xlabel('False Positive Rate', fontsize=18)
plt.ylabel('True Positive Rate', fontsize=18)
plt.title('Receiver operating characteristic for predicting high salary', fontsize=18)
plt.legend(loc="lower right")
plt.show()
```

The AUC ROC value is about 0.798, which is fairly high. I conclude that the tweaked model is still fairly robust, even after lowering the number of false positives to 0.
